```python
#change column from text to number values

from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()


COLUMN_NAME = "State"

#print original unique values in column
print("unique values:",df[COLUMN_NAME].unique())

#text_column = np.array(df['Region'])
text_column = np.array(df[COLUMN_NAME]).reshape(-1, 1)

encoder.fit(text_column)

text_column_encoded = encoder.transform(text_column)

print("column",text_column_encoded)

#get unique values
print("unique values:",np.unique(text_column_encoded))
```
