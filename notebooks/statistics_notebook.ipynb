```python
#change column from text to number values



def translate_string_column_to_int(the_column_name):
    from sklearn.preprocessing import OrdinalEncoder
    encoder = OrdinalEncoder()


    COLUMN_NAME = the_column_name

    #print original unique values in column
    print("unique values:",df[COLUMN_NAME].unique())

    #text_column = np.array(df['Region'])
    text_column = np.array(df[COLUMN_NAME]).reshape(-1, 1)

    encoder.fit(text_column)

    text_column_encoded = encoder.transform(text_column)

    #print("column",text_column_encoded)

    #get unique values
    print("unique values:",np.unique(text_column_encoded))
    
    return text_column_encoded


ddd = translate_string_column_to_int("Industry")

print("Industry val:",np.unique(ddd))
```

```python
#translate column names
non_translatable_columns = ["PeopleInHousehold","MetroAreaCode","Age","Hispanic","CountryOfBirthCode"]
translate_column_names = ["Region","State","Married","Sex","Education","Race","Citizenship","EmploymentStatus","Industry"]

#make new rows with numbers
df2 = df.copy()

for i in translate_column_names:
    df2[i+"_num"]=translate_string_column_to_int(i)
    print("unique values "+i+ " new:",df2[i+"_num"].unique())

#make a csv file for explanation:
#df2.to_csv('file_name.csv')

# delete the replaced columns
for z in translate_column_names:
    df2.drop([z],1,inplace=True)

# print columns
df2.info()
```

```python
#rearrange columns

new_order = ["PeopleInHousehold","Region_num","State_num","MetroAreaCode","Age","Married_num","Sex_num","Education_num","Race_num","Hispanic","CountryOfBirthCode","Citizenship_num","EmploymentStatus_num","Industry_num"]


def order(frame,var):
    if type(var) is str:
        var = [var] #let the command take a string or list
    varlist =[w for w in frame.columns if w not in var]
    frame = frame[var+varlist]
    return frame

df2 = order(df2,new_order)

df2.info()
```
