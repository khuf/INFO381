```python
import pandas as p
import numpy as np
from sklearn.feature_extraction import DictVectorizer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
cps = p.read_csv("../data/CPSData.csv")
```

```python
#remove all rows with NAN values
cps = cps.dropna()
```

```python
vec = DictVectorizer()
```

```python
cps.Education.value_counts()
```

```python
cps
```

```python
#add age by decade column
cps['age_by_decade'] = p.cut(x=cps['Age'], bins=[10,20,29,39,49,59,69,79,89,99], labels=[10,20,30,40,50,60,70,80,90])
```

```python
cps
```

```python
cps.iloc[0]
```

```python
#transform unique values to rows

ohencoder = OneHotEncoder(dtype=np.int, sparse=True)

nominals = p.DataFrame(
    ohencoder.fit_transform(cps[['EmploymentStatus','Education','age_by_decade']])\
    .toarray(),
    columns=np.concatenate((cps["EmploymentStatus"].unique(),cps["Education"].unique(),cps["age_by_decade"].unique()), axis=None))
#cps.head()
nominals
```
